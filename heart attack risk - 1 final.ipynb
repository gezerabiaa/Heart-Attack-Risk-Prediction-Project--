```python
numeric_df = df.select_dtypes(include=['number'])
cor_matrix = numeric_df.corr().abs()

#Eğer veri kümenizde 'BMW7812' gibi string veriler bulunuyorsa, bu verileri içermeyen bir alt küme oluşturarak korelasyon 
#matrisi oluşturabilirsiniz. Örneğin, sadece sayısal sütunları içeren bir alt küme oluşturarak şöyle bir şey yapabilirsiniz:
```

```python
import numpy as np
upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape) , k=1).astype(bool))
drop_list = [col for col in upper_triangle_matrix.columns if any (upper_triangle_matrix[col]>0.90)] #korelasyonu yüksek olanlarla çalışmak istiyorum 
cor_matrix[drop_list]

# yüksek korelasyona sahip olan sayısal değişkenleri belirlemek için bir işlem gerçekleştirir. 
```

```python
df = df.drop(drop_list , axis=1)
sensor = sensor.drop(drop_list, axis=1) 
#verisetinden yoğun korele değerleri siliyorum

#Bu kod parçası, yüksek korelasyona sahip olan değişkenleri df ve sensor DataFrame'lerinden çıkarır. Bu, modelleme sürecinde gereksiz değişkenleri eleme veya daha az sayıda değişkenle çalışma ihtiyacından kaynaklanabilir.
```

```python
#optimum küme sayısını belirliyorum
#clustiring kümeleme( bir sürü bbenzer datayı aynı yerde toplama)
#clustriign unsupervised taxonomysinde bulunuyor
# Bu, k-ortalama (k-means) kümeleme algoritmasıyla gerçekleştirilir ve belirli bir aralıktaki k değerleri için çözümleri değerlendirerek kümeleme performansını anlamayı amaçlar. 

kmeans=KMeans()
elbow = KElbowVisualizer(kmeans, k=(2,20))
elbow.fit(X_scaled_v1)
elbow.show()
```

```python
elbow.elbow_value_ #8 grup çıktı 8 cluster var 
```

```python
kmeans = KMeans(n_clusters = elbow.elbow_value_, random_state=17).fit(X_scaled_v1)
kmeans.get_params()

# 8 gruubu parametre değerlerini çalıştırarak inceledim
 #bu kümeleme modelinin parametre değerlerini inceledim 

```

```python
kmeans.n_clusters  #küme sayısı 
kmeans.cluster_centers_ #küme merkezleri 
kmeans.labels_
kmeans.inertia_ #sabit değer

#benim için öenmli olanlar 8 cluster'ın merkezleri ,labellarım  , init diye bir sabit değerim
#126700.83150298812  sabit değer 
```
